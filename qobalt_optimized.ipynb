```python
def getMinimumTime(requestedServers, transitionTime):
    n = len(transitionTime)
    if n == 0:
        return 0

    # prefix[i] = total clockwise cost to move from server 1 to server i + 1.
    prefix = [0] * (n + 1)
    for i, cost in enumerate(transitionTime, start=1):
        prefix[i] = prefix[i - 1] + cost

    ring_total = prefix[n]
    total_time = 0
    current = 1

    for target in requestedServers:
        if not 1 <= target <= n:
            raise ValueError(f"server {target} is out of range 1..{n}")

        if target == current:
            continue

        a = current - 1
        b = target - 1

        if a <= b:
            clockwise = prefix[b] - prefix[a]
        else:
            clockwise = ring_total - (prefix[a] - prefix[b])

        counterclockwise = ring_total - clockwise
        total_time += min(clockwise, counterclockwise)
        current = target

    return total_time


# Example:
# requestedServers = [3, 2, 3]
# transitionTime = [2, 1, 4]
# print(getMinimumTime(requestedServers, transitionTime))
```
