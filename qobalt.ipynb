```python
def getMinimumTime(requestedServers, transitionTime):
    m = len(transitionTime)
    total_time = 0
    current_server = 1  # assuming servers are 1-indexed

    for target in requestedServers:
        if target == current_server:
            continue

        # Compute clockwise time
        clockwise_time = 0
        i = current_server
        while i != target:
            clockwise_time += transitionTime[(i - 1) % m]
            i = (i % m) + 1

        # Compute counterclockwise time
        counterclockwise_time = 0
        i = current_server
        while i != target:
            i = (i - 2) % m + 1
            counterclockwise_time += transitionTime[(i - 1) % m]

        # Choose the minimum time
        total_time += min(clockwise_time, counterclockwise_time)
        current_server = target

    return total_time
```
